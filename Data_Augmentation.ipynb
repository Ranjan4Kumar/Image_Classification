```python
import tensorflow as tf
from tensorflow import keras
import keras.utils as image # # this work as an open cv because it is inside keras
from keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('/content/drive/MyDrive/DS/TensorFlow_img_Classification_Project/cat.10.jpg',target_size=(200,200))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
type(img) # Here we have image type(PIL -> Python imaging library) but when we use open CV it is in the form of array
```

```python
# We created an object whihc will tell us which kind or images we have to create

datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        rescale=1./255,
        shear_range=0.2,      
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
# In our module " image" we have one function called imgae to array to convert image to an array

img = image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3) # this dimessons means 1 image of 200X200X3 
```

```python
i = 0

# Here we have flow function which is used when we are passing only one image for genersting multiple images. flow_from_directory(used when we are passing multiple images from a directory)
for output in datagen.flow(input_batch, batch_size = 1,save_to_dir = '/content/drive/MyDrive/DS/TensorFlow_img_Classification_Project/aug'):
  i = i+1

  if i==50:
    break

# we are creating only 10 images
```
